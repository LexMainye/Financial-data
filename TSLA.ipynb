```python
# import pandas , datareader & date
import pandas as pd 
from pandas_datareader.data import DataReader
from datetime import date
# enter startdate
start = date(2011,1,1)
#enter enddate
end = date(2021,3,2)#today's data
#enter ticker
ticker = 'TSLA'
#enter datasource
data_source = 'yahoo'
#create tsla data 
stock_data = DataReader(ticker, data_source,start,end)
#check out tsla data
print(stock_data.head())
stock_data.info()
#concact tsla data
pd.concat([stock_data.head(3),stock_data.tail(3)])
```

```python
#import matplotlib
import matplotlib.pyplot as plt
stock_data['Close'].plot(title=ticker)
plt.show()
```
